# Chechking weird values

```python
def try_parse(n):
  try:
    return float(n)
  except:
    return 0 #-inf would make nsmallest useless

for var in variables:
  parsed_series = df[var].apply(try_parse)
  print(var.upper() + "\n")
  print("MAX")
  print(parsed_series.nlargest(5))
  print("MIN")
  print(parsed_series.nsmallest(5))
  print()
```

We will make all weird values null


```python
def parse(n, var):
  try:
    parsed_val = float(n)
    match var:
      case "ta_2m":
        return parsed_val if -50 <= parsed_val <= 50 else "null"
      case "pa":
        return parsed_val if 750 <= parsed_val <= 1080 else "null"
      case "rh":
        return parsed_val if 0 <= parsed_val <= 120 else "null"
      case "pr_1h":
        return parsed_val if 0 <= parsed_val <= 500 else "null" #no news about extreme rainfall in Jakubovany
      case "ws_avg":
        return parsed_val if 0 <= parsed_val <= 300 else "null" #no news about extreme wind in Javorniky
      case "wd_avg":
        return parsed_val if 0 <= parsed_val <= 360 else "null" #no other possible values
  except:
    return "null"
```

```python
dataframes = {}
for var in variables:
    df_var = df.pivot(index="date", columns="cccc", values=var)
    df_var = df_var.map(lambda n: parse(n, var))
    dataframes[var] = df_var
```

```python
import shutil
for var, dataframe in dataframes.items():
    dataframe.to_csv(f"output/{var}_data.csv")

shutil.make_archive("data", 'zip', "output")
```
